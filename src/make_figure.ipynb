```python
import pandas as pd
import csv

data = {
    'Method': [],
    'Sessions': [],
    'PixCorr': [],
    'SSIM': [],
    'AlexNet(2)': [],
    'AlexNet(5)': [],
    'InceptionV3': [],
    'CLIP': [],
    'EffNet-B': [],
    'SwAV': [],
    'FwdRetrieval': [],
    'BwdRetrieval': [],
    'Brain Corr. nsd_general': [],
    'Brain Corr. V1': [],
    'Brain Corr. V2': [],
    'Brain Corr. V3': [],
    'Brain Corr. V4': [],
    'Brain Corr. higher_vis': []}

df = pd.DataFrame(data)
files = ["brain_diffuser_125_subj01.csv", "brain_diffuser_375_subj01.csv", "brain_diffuser_750_subj01.csv", "brain_diffuser_1500_subj01.csv", "brain_diffuser_15000_subj01.csv", "brain_diffuser_27750_subj01.csv", "brain_diffuser_30000_subj01.csv",
         "takagi_decoded_text_01_subj01.csv", "takagi_decoded_text_05_subj01.csv", "takagi_decoded_text_1_subj01.csv", "takagi_decoded_text_2_subj01.csv", "takagi_decoded_text_20_subj01.csv", "takagi_decoded_text_37_subj01.csv", "takagi_decoded_text_40_subj01.csv",
         "mindeye2_125_pt_subj01.csv", "mindeye2_375_pt_subj01.csv", "mindeye2_1_pt_subj01.csv", "mindeye2_2_pt_subj01.csv", "mindeye2_20_pt_subj01.csv", "mindeye2_37_pt_subj01.csv", "mindeye2_40_pt_subj01.csv",
         #"mindeye2_125_subj01.csv", "mindeye2_375_subj01.csv", 
         "mindeye2_1_subj01.csv", "mindeye2_2_subj01.csv", "mindeye2_20_subj01.csv", "mindeye2_37_subj01.csv", "mindeye2_40_subj01.csv",]
methods = ["brain_diffuser", "brain_diffuser", "brain_diffuser", "brain_diffuser", "brain_diffuser", "brain_diffuser", "brain_diffuser",
           "takai_decoded_text", "takai_decoded_text", "takai_decoded_text", "takai_decoded_text", "takai_decoded_text", "takai_decoded_text", "takai_decoded_text",
           "mindeye2_pt", "mindeye2_pt", "mindeye2_pt", "mindeye2_pt", "mindeye2_pt", "mindeye2_pt", "mindeye2_pt", 
        #    "mindeye2", "mindeye2", 
           "mindeye2", "mindeye2", "mindeye2", "mindeye2", "mindeye2", "mindeye2"]
sessions = [0.166, 0.5, 1, 2, 20, 37, 40, 0.166, 0.5, 1, 2, 20, 37, 40, 0.166, 0.5, 1, 2, 20, 37, 40, 
            # 0.166, 0.5, 
            1, 2, 20, 37, 40]
for file, method, sessions in zip(files, methods, sessions):
    data_list = []
    with open(f'tables/{file}', 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            row_values = [float(value) if value != "" else None for value in row]
            data_list.append(row_values)
    new_row_data = {
    'Method': method,
    'Sessions': sessions,
    'PixCorr': data_list[0],
    'SSIM': data_list[1],
    'AlexNet(2)': data_list[2],
    'AlexNet(5)': data_list[3],
    'InceptionV3': data_list[4],
    'CLIP': data_list[5],
    'EffNet-B': data_list[6],
    'SwAV': data_list[7],
    'FwdRetrieval': data_list[8],
    'BwdRetrieval': data_list[9],
    'Brain Corr. nsd_general': data_list[10],
    'Brain Corr. V1': data_list[11],
    'Brain Corr. V2': data_list[12],
    'Brain Corr. V3': data_list[13],
    'Brain Corr. V4': data_list[14],
    'Brain Corr. higher_vis': data_list[15]}
    df = pd.concat([df, pd.DataFrame(new_row_data)])
print(df)
df.to_csv('tables/output.csv', index=False)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = df
# Setting the plot style
sns.set(style="whitegrid")

# Getting the list of methods and metrics
methods = data['Method'].unique()
# metrics = data.columns[2:]

data['Sessions_str'] = data['Sessions'].astype(str)

# Creating a figure and a set of subplots
fig, ax = plt.subplots(figsize=(15, 8))
line_styles = ['-', '--', '-.', ':']
selected_metrics = ['InceptionV3', 'Brain Corr. nsd_general', 'CLIP', "SSIM"]
bright_palette = sns.color_palette("bright", len(methods))
colors = ["red", "yellow", "blue", "green"]
# Normalizing the selected metrics
normalized_data = data.copy()
for metric in selected_metrics:
    normalized_data[metric] = (normalized_data[metric] - normalized_data[metric].min()) / (normalized_data[metric].max() - normalized_data[metric].min())
# metric = 'Brain Corr. nsd_general'
for method_idx, method in enumerate(methods):
    method_data = normalized_data[normalized_data['Method'] == method]
    for metric_idx, metric in enumerate(selected_metrics):
        sns.lineplot(x='Sessions_str', y=metric, data=method_data, ax=ax, 
                     color=colors[method_idx], label=f"{method} - {metric}",
                     linestyle=line_styles[metric_idx])

# Enhancing the plot
# Setting meaningful x-ticks and y-ticks
# Setting custom x-ticks and y-ticks
# x_ticks = [0.166, 0.5, 1, 2, 20, 37, 40]
x_tick_labels = ["10 minutes", "30 minutes", "1 hour", "2 hours", "20 hours", "37 hours", "40 hours"]
ax.set_xticklabels(x_tick_labels, rotation=45)

# Setting y-ticks
y_ticks = np.linspace(data[metric].min(), data[metric].max(), num=10)
ax.set_yticks(y_ticks)
ax.set_yticklabels([f"{tick:.2f}" for tick in y_ticks])

ax.set_xlabel('Number of Sessions', fontsize=14)
ax.set_ylabel('Normalized Metric Values', fontsize=14)
ax.set_title('Performance of Selected Metrics Across Different Methods (Normalized)', fontsize=16)
ax.legend(title='Method and Metric', loc='upper left', bbox_to_anchor=(1, 1))
plt.yscale('linear')
plt.tight_layout()

# Show the plot
plt.show()
```
